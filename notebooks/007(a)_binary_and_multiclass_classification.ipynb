**test_size=0.2**: This specifies the split ratio. Here, 20% of your data will be randomly set aside for the test set, and the remaining 80% will be used to train the model.

 > What does stratify=y mean? (Crucial Step)

This is a powerful argument, especially for classification problems. Stratification ensures that the train and test sets have the exact same proportion of class labels as the original dataset.

Imagine you have a dataset of 100 patients where 80 are healthy (0) and 20 have a disease (1).

    Without stratification: A purely random split might accidentally put all 20 diseased patients into the training set, leaving 0 in the test set. Your test results would be completely useless.

    With stratify=y: Pandas forces the split to maintain that 80/20 balance. Both your X_train and X_test will consist of exactly 80% healthy patients and 20% diseased patients.

```python
zero_as_missing = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

df[zero_as_missing] = df[zero_as_missing].replace(0, np.nan)
df[zero_as_missing] = df[zero_as_missing].fillna(df[zero_as_missing].median())

X = df.drop('Outcome', axis=1)
y = df['Outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=RANDOM_SEED, stratify=y)

# local model storage
model_dir = "D:/CitrusBits/pythonic-rebirth/models"
os.makedirs(model_dir, exist_ok=True)
dt_path = os.path.join(model_dir, "indian_diabetes_decision_tree.pkl")

# load or train again
if os.path.exists(dt_path):
    print(f"Found existing model at {dt_path}, loading instead of retraining...")
    clf = joblib.load(dt_path)
else:
    print("No saved model found, training new Decision Tree...")
    clf = DecisionTreeClassifier(
        max_depth=4,
        min_samples_leaf=5,
        criterion='gini',
        random_state=42
    )
    clf.fit(X_train, y_train)
    joblib.dump(clf, dt_path)
    print(f"Model saved to {dt_path}")

# evaluate
y_pred = clf.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))

```

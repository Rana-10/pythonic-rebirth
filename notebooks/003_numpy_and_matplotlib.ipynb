**Q: Boolean Masking**
* Filter values in arrays using conditions.
* Combine multiple conditions with & (and), | (or).
* Count and sum filtered elements.


**Boolean masking** is a technique used to filter elements of an array based on one or more conditions.

Instead of looping through the array, NumPy creates a Boolean array (True and False) and returns only the elements where the condition is True.

```python
arr = np.array([10, 20, 30, 40, 50])

mask = arr > 25
print(mask)

print(arr[arr > 25])

arr_2 = np.array([
    [1, 2, 3],
    [4, 7, 9],
    [10, 12, 14]
])

print(arr_2[arr_2 > 8])

print(arr_2[(arr_2 > 10) & (arr_2 < 15)])

# to count filtered elements we use
# len() or np.count_nonzero()

print(len(arr[arr > 25]))
print(np.count_nonzero(arr > 25))

print(np.sum(arr[arr > 25]))
```

**Q: Plotting with Matplotlib**
* Import matplotlib.pyplot as plt.
* Create line plots using plt.plot().
* Create bar plots using plt.bar().
* Create scatter plots using plt.scatter().
* Create histograms using plt.hist().


To make a good plot from Matplotlib, keep this shortcut in mind
TXYLGS for Title, X-axis, Y-axis, legend, grid and style

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 20, 18]

# line plot
plt.plot(x, y,
         color="blue",
         linestyle="--",
         marker="o",
         linewidth=2,
         label="Sales")

plt.grid(True)
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales $")

plt.show()

plt.plot(x, y,
         color='red',
         linestyle='--',
         marker='o',
         linewidth=2)

plt.title("Customized Line Plot")
plt.xlabel("Days")
plt.ylabel("Sales")

plt.show()

# bar plot
subjects = ["Math", "Physics", "Chemistry", "English"]
marks = [85, 78, 90, 88]

plt.bar(subjects, marks)

plt.title("Student Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")

plt.show()

# scatter plot
height = [150, 108, 160, 165, 170]
weight = [45, 50, 58, 48, 72]

plt.scatter(height, weight)

plt.title("Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")

plt.show()

# histogram
marks = [45, 55, 60, 62, 67, 70, 71, 72, 75, 78,
         80, 82, 85, 88, 90, 92, 95]

plt.hist(marks)

plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.show()
```

**Q: Array Creation and Properties
Create 1D, 2D, and 3D arrays using np.array, np.zeros, np.ones, and np.full.
Use np.arange() and np.linspace() to create arrays with evenly spaced values.
Check array properties: .ndim, .shape, .size, .dtype.**


np.arange(start, stop, step)
* Works like Python's range(), but returns an array. Stop is exclusive.

np.linspace(start, stop, num)
* Generates num evenly spaced values. Stop is inclusive by default.

```python
import numpy as np

# making 1D, 2D AND 3D Arrays

arr_1d = np.array([1, 2, 4, 6, 10])

arr_2d = np.array([[1, 4, 7], [2, 5, 8]])

arr_3d = np.array([[[1, 3], [5, 6]],
                   [[7, 8], [9, 0]]])

# using np.zeros() np.ones() np.full() to populate

zeros_1d = np.zeros(5)
zeros_2d = np.zeros((4, 5))
# print(zeros_2d)

zeros_3d = np.zeros((2, 3, 1))  # 2 blocks of 3x1
# print(zeros_3d)

full_1d = np.full(5, 7)
#print(full_1d)

full_2d = np.full((3, 4), 5)  # 3x4 matric all elements =5
# print(full_2d)

full_3d = np.full((2, 3, 4), 6)
# print(full_3d)


a_matrix = np.arange(0, 10, 2)
# print(a_matrix)

b_matrix = np.arange(6)
# print(a_matrix)

c_matrix = np.arange(1, 2, 0.25)
# print(c_matrix)

e_matrix = np.linspace(0, 10, 5)
# print(e_matrix)


# checking array properties

print(e_matrix.ndim)
print(e_matrix.shape)  # in Python, a tuple with one element is written as (value,)
print(e_matrix.size)
# print(e_matrix.dtype)
```

Every estimator has:

* fit(X, y) -> learn from data
* predict(X) -> make predictions
* transform(X) -> for preprocessing/feature transformers
* fit_transform(X) -> fit + transform in one call (used on training data only)

 **Note:** `train_test_split()` always returns **two datasets**. The first contains the remaining data after the split, and the second contains the portion specified by `test_size` (or `train_size`). The variable names (e.g., `train_val`, `test`, `train`, `val`) are **user-defined** and can be anything.

```python
from sklearn.model_selection import train_test_split

# splitting the sets into 60,30,10 ratio

# train_val means training+validation
train_val, test = train_test_split(dataset, test_size=0.30, shuffle=True, random_state=42)

train, val = train_test_split(train_val, test_size=(0.10 / 0.70), shuffle=True, random_state=42)

print(f"Train shape: {train.shape}")
print(f"Validation shape: {val.shape}")
print(f"Test shape: {test.shape}")
```

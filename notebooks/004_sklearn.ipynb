**Q: Split the data into test (30%), train (60%), and validation (10%). Randomize/Shuffle the data and visualize it.**

 **Note:** `train_test_split()` always returns **two datasets**. The first contains the remaining data after the split, and the second contains the portion specified by `test_size` (or `train_size`). The variable names (e.g., `train_val`, `test`, `train`, `val`) are user-defined and can be anything.

```python
from sklearn.model_selection import train_test_split

# splitting the sets into 60,30,10 ratio

# train_val means training+validation
train_val, test = train_test_split(dataset, test_size=0.30, shuffle=True, random_state=42)

train, val = train_test_split(train_val, test_size=(0.10 / 0.70), shuffle=True, random_state=42)

print(f"Train shape: {train.shape}")
print(f"Validation shape: {val.shape}")
print(f"Test shape: {test.shape}")
```

**Q: Use StandardScaler and MinMaxScaler from sklearn.preprocessing to scale the numerical features (e.g., age, bmi, charges).
Visualize the effect of both scaling techniques using histograms or boxplots.**


**StandardScaler** is also called Z-score normalization. It rescales so that mean = 0 and std = 1. Values can go negative, and the range is not bounded. It just shifts/rescales around the center of the distribution. Good when your data roughly follows a normal distribution or when outliers shouldn't dominate the scale.

**MinMaxScaler** compresses everything into a fixed range (0 to 1 by default) using (x - min) / (max - min). Every value ends up bounded, but it's very sensitive to outliers. One extreme value can stretch the whole scale.

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# select the numerical columns you want to scale
numerical_cols = ['age', 'bmi', 'charges']

# standardscaler
standard_scaler = StandardScaler()
scaled_standard = standard_scaler.fit_transform(dataset[numerical_cols])
df_standard_scaler = pd.DataFrame(scaled_standard, columns=numerical_cols)

# minmaxScaler
minmax_scaler = MinMaxScaler()
scaled_minmax = minmax_scaler.fit_transform(dataset[numerical_cols])
df_minmax_scaler = pd.DataFrame(scaled_minmax, columns=numerical_cols)

print("Original stats:\n", dataset[numerical_cols].describe())
print("\nStandardScaler stats:\n", df_standard_scaler.describe())
print("\nMinMaxScaler stats:\n", df_minmax_scaler.describe())
```

**Q: The dataset has categorical columns like sex, smoker, and region.
Use LabelEncoder and OneHotEncoder from sklearn.preprocessing to encode these features.**

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# LabelEncoder for binary categorical columns
label_encoder = LabelEncoder()

dataset['sex_encoded'] = label_encoder.fit_transform(dataset['sex'])
print("sex classes:", label_encoder.classes_)  # shows which label = 0, which = 1

dataset['smoker_encoded'] = label_encoder.fit_transform(dataset['smoker'])
print("smoker classes:", label_encoder.classes_)

print(dataset[['sex', 'sex_encoded', 'smoker', 'smoker_encoded']].head())
```

```python
# OneHotEncoder for multi-category columns
onehot_encoder = OneHotEncoder(sparse_output=False)  # use sparse=False if on older sklearn

region_encoded = onehot_encoder.fit_transform(dataset[['region']])

# convert to DataFrame with proper column names
region_encoded_df = pd.DataFrame(
    region_encoded,
    columns=onehot_encoder.get_feature_names_out(['region'])
)

print(region_encoded_df.head())

# combine back with original dataset
dataset = pd.concat([dataset, region_encoded_df], axis=1)
print(dataset.head())
```

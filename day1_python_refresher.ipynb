Q: Write a function to check if a given string is a palindrome (ignoring spaces and case). word, phrase, or sequence that reads the same backwards as forwards, e.g. madam or nurses run.)

```python
def is_palindrome(text: str):
    clean = ""

    for ch in text:  # removing space + making lowercase
        if ch != " ":
            clean += ch.lower()

    start = 0
    end = len(clean) - 1

    while start < end:
        if clean[start] != clean[end]:
            return False
        start += 1
        end -= 1

    return True


if (is_palindrome("civic") == True):
    print("The number is palindromic")
else:
    print("The number is not palindromic")

```

Q: Given a 2D matrix, return its transpose without using external libraries.

```python
a: list[list[int]] = [[0 for _ in range(3)] for _ in range(3)]

count: int = 1
for i in range(0, 3):
    for j in range(0, 3):
        a[i][j] = count
        count += 1

print("Printing original matrix:")
for i in range(0, 3):
    for j in range(0, 3):
        print(a[i][j], end=" ")
    print()

# swapping, doing transpose
for i in range(0, 3):
    for j in range(i + 1, 3):
        a[i][j], a[j][i] = a[j][i], a[i][j]

print("Printing transposed matrix:")
for i in range(0, 3):
    for j in range(0, 3):
        print(a[i][j], end=" ")
    print()
```

Q: Write a function that returns True if a number is prime, else False.

```python
def prime_num(n: int):
    count: int = 0
    for i in range(2, n + 1, 1):
        if n % i == 0:
            count += 1
    if count == 1:
        return True
    else:
        return False


if (prime_num(29) == True):
    print("It is a prime number")
else:
    print("It is not a prime number")
```

Q: Create a Contact class with name, phone, and email. Create a ContactBook class to store multiple contacts and search by name

```python
class Contact:
    def __init__(self, name, phone, email):
        self.name = name
        self.phone = phone
        self.email = email


class ContactBook:
    def __init__(self):
        self.contacts = []

    def add_contact(self, contact):
        self.contacts.append(contact)

    def search_by_name(self, name):
        for i in self.contacts:
            if i.name == name:
                return i


c1 = Contact("Ali Hamza", "03237653812", "alihamza2@gmail.com")
c2 = Contact("Sara Qamar", "03209866554", "saraq2@gmail.com")

book = ContactBook()
book.add_contact(c1)
book.add_contact(c2)

result = book.search_by_name("Sara Qamar")

if result:
    print("Person's phone number is ", result.phone, "and email is",
          result.email)
else:
    print("Not found")

```

Q: Class Item (name, price), class Cart with methods to add_item, remove_item, total_cost, and list_items.

```python
class Item:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __str__(self):
        return self.name + " - " + str(self.price)


class Cart:
    def __init__(self):
        self.items = []

    def add_item(self, itemm):
        self.items.append(itemm)

    def remove_item(self, itemm):
        self.items.remove(itemm)

    def total_cost(self):
        total = 0
        for i in self.items:
            total = total + i.price
        return total

    def show_list_items(self):
        for item in self.items:
            print(item)


item1 = Item("Tekken 8", 60)
item2 = Item("Battlefield 6", 59)

cart1 = Cart()
cart1.add_item(item1)
cart1.add_item(item2)

cart1.show_list_items()
print()
print("The total cost currently is", cart1.total_cost())

print()

cart1.remove_item(item2)

cart1.show_list_items()

print()

print("The total cost currently is", cart1.total_cost())
```

Q: Write a function that generates the first n numbers in the Fibonacci sequence.

```python
def fibonacci_seq(n: int):
    first_num = 0
    second_num = 1
    temp = 0

    fibo = []

    for i in range(n):
        fibo.append(first_num)
        temp = first_num + second_num
        first_num = second_num
        second_num = temp

    return fibo


print(fibonacci_seq(6))
```

Q: Question class with text and answer, Quiz class with list of questions and method to run the
quiz (text-based).

```python
class Question:
    def __init__(self, text, answer):
        self.text = text
        self.answer = answer


class Quiz:
    def __init__(self, questions):
        self.questions = questions

    def run_quiz(self):
        score = 0

        for question in self.questions:
            print(question.text)

            user_answer = input("Your answer: ")

            if user_answer.lower() == question.answer.lower():
                score += 1
                print("Correct!\n")
            else:
                print("Wrong!\n")

        print("Your score:", score, "/", len(self.questions))


q1 = Question("What is the capital of Pakistan?", "Islamabad")
q2 = Question("What is 2 + 2?", "4")
q3 = Question("Which language are we learning?", "Python")

quiz1 = Quiz([q1, q2, q3])

quiz1.run_quiz()
```

Q: Base class Shape with method area(), subclasses: Triangle, Circle, Rectangle. Demonstrate polymorphism via a list of shapes

```python
import math


class Shape:
    def area(self):
        pass


class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height


class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius * self.radius


shapes = [
    Triangle(10, 5),
    Rectangle(4, 6),
    Circle(3)
]

for shape in shapes:
    print("Area:", shape.area())
```

Q: Given a .txt file path, create a class to read the file and return number of lines, number of words, and most frequent word

```python
class FileReader:
    def __init__(self, file_path):
        self.file_path = file_path

    def read_file(self):
        file = open(self.file_path, "r")
        data = file.read()
        file.close()

        return data

    def number_of_lines(self):
        file = open(self.file_path, "r")
        lines = file.readlines()
        file.close()

        return len(lines)

    def number_of_words(self):
        data = self.read_file()

        words = data.split()

        return len(words)

    def most_frequent_word(self):
        data = self.read_file()

        words = data.split()

        frequency = {}

        for word in words:
            if word in frequency:
                frequency[word] += 1
            else:
                frequency[word] = 1

        most_common = ""
        max_count = 0

        for word in frequency:
            if frequency[word] > max_count:
                max_count = frequency[word]
                most_common = word

        return most_common


# testing using an example
file1 = FileReader("example")

print("Number of lines:", file1.number_of_lines())
print("Number of words:", file1.number_of_words())
print("Most frequent word:", file1.most_frequent_word())
```

Q: Check if two strings are anagrams of each other.
(a word, phrase, or name formed by rearranging the letters of another, such as spar, formed
from rasp.)

```python
def is_anagram(str1: str, str2: str):
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()

    return sorted(str1) == sorted(str2)


print(is_anagram("civic", "civic"))
print(is_anagram("hello", "world"))
```

### pip
It is the default, built in Python package installer, and comes with Python automatically.
Basically, think of it like an app store where you search, it installs.

Pros:
Comes pre-installed with Python, zero setup
It is simple and straightforward

Cons:
No built-in virtual environment management
No dependency conflict resolution, it installs whatever you ask, even if versions clash
No lock file by default which means it will be hard to replicate exact environment on another machine

### Anaconda

Anaconda is a full distribution, not just a package manager, but also comes with Python. It has 250+ pre-installed scientific libraries, and its own environment manager.
Where it separates itself is that it is built specifically for data science and scientific computing. Manages both Python and non-Python packages (like C libraries that numpy/scipy depend on under the hood).
We can imagine it like buying a laptop that already has all your software pre-installed.


Pros:
Handles non-Python dependencies (C, R, CUDA, etc.)which is huge for ML/DL
Excellent virtual environment management
Pre-installed with Jupyter, numpy, pandas, matplotlib out of the box
Avoids DLL/binary issues on Windows (common headache with pip)

Cons:
Very heavy to install (several GBs)
Slower than pip
Can mix poorly with pip if you use both carelessly


### Poetry
Poetry is a modern Python dependency manager and packaging tool, and not a Python distribution. It just manages packages and virtual environments cleanly.
Purpose: Professional project management — tracks exact versions, resolves conflicts, and makes your project reproducible on any machine.
Analogy: Like a project manager who keeps a detailed logbook of every decision made.

Pros:
Lock file (poetry.lock) guarantees identical environments everywhere
Clean virtual environment per project
Professional standard for production and collaborative projects

Cons:
Extra setup required as it is not built into Python
Slight learning curve
Slower adoption in data science community compared to ML/research communities
